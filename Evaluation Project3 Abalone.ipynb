```python
#Treating skewness
import numpy as np
for col in df_abalone.columns:
    if df_abalone.skew().loc[col]>0.55:
        df_abalone[col]=np.log1p(df_abalone[col])
```

```python
df_abalone.skew()
```

```python
#Checking outliers
from scipy.stats import zscore
z_score=abs(zscore(df_abalone))
print(df_abalone.shape)
df_abalone_final=df_abalone.loc[(z_score<3).all(axis=1)]
print(df_abalone_final.shape)
```

```python
#Separate target and input variables
import numpy as np
df_x=df_abalone_final.drop(columns=["age"])
y=df_abalone_final[["age"]]
```

```python
#Scaleinput variables for all features to be on common scale in linear regression
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(df_x)
x=pd.DataFrame(x,columns=df_x.columns)
```

```python
#check skewness again
x.skew()
```

```python
#calculating max r2 score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model
max_r_score=0
for r_state in range(42,100):
    x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = r_state,test_size=0.20)
    regr = linear_model.LinearRegression()
    regr.fit(x_train,y_train)
    y_pred = regr.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
print("max r2 score corresponding to ",final_r_state," is ",max_r_score)
```

```python
#Cross validation
from sklearn.model_selection import cross_val_score
cross_val_score(linear_model.LinearRegression(),x,y,cv=10,scoring="r2").mean()
```

```python
#max r2 score is at random state 50
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state = 50,test_size=0.20)
lreg = linear_model.LinearRegression()
lreg.fit(x_train,y_train)
y_pred=lreg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
print("r2 score is: ",r2_score(y_test,y_pred))
print("RMSE is : ",np.sqrt(mean_squared_error(y_test,y_pred)))
```
